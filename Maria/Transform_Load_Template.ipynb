```python
#analyze data #####################

rating_vs_sales = pd.read_sql('select IMDB, SUM(`Total Sales`) as Total_Sales from imdbdisney i \
                    left join disneygrosssales g on i.Title = g.Title \
                    group by IMDB order by SUM(`Total Sales`) desc',conn)
rating_vs_sales['Total_Sales'] = rating_vs_sales['Total_Sales'].map('${:0,.2f}'.format)
rating_vs_sales.head(5)

#FUN FACT = as expected the highest rating movies has the highest total sales
```

```python
years_vs_sales = pd.read_sql('select year, `Total Sales` from disneygrosssales order by `Total Sales` desc;',conn)
years_vs_sales['Total Sales'] = years_vs_sales['Total Sales'].map('${:0,.2f}'.format)
years_vs_sales.head(5)


#NOTE = we need to expand this dataset as the results are not enough to conclude if year is related to total sales.
#PER RESULT BELOW: the highest total sales are not related with years
```

```python
actor_voice_movie  = pd.read_sql('select voice_actor, count(distinct movie) as count_of_movies, \
                    count(distinct `character`) as count_of_characters \
                    from dis_voices group by voice_actor order by count(distinct movie) desc', conn)
actor_voice_movie.head(10)

#FUN FACT = When analyzing the involvement of voice actors per movie, each voice actor 
#generally played multiple roles in a movie.

#test with exameple
#actor_voice_new = pd.read_sql('select voice_actor,movie,`character` \
#                    from dis_voices where voice_actor = "Frank Welker" order by movie', conn)
#actor_voice_new.head(5)
```
